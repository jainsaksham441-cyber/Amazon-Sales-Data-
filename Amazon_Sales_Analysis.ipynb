```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Amazon.csv")
df
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)

```

```python
df.dropna(inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
df['Year'] = df['OrderDate'].dt.year
df['Month'] = df['OrderDate'].dt.month
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Revenue'] = df['Quantity'] * df['UnitPrice']
df['Final_Sales'] = df['Revenue'] - df['Discount']
```

```python
df['Final_Sales'].sum()
```

```python
category_sales = df.groupby('Category')['Final_Sales'].sum().sort_values(ascending=False)
category_sales

```

```python
top_products = df.groupby('ProductName')['Final_Sales'].sum().sort_values(ascending=False).head(10)
top_products
```

```python
monthly_sales = df.groupby('Month')['Final_Sales'].sum()
monthly_sales
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=category_sales.index, y=category_sales.values)
plt.xticks(rotation=45)
plt.title("Sales by Category")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title("Monthly Sales Trend")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=top_products.values, y=top_products.index)
plt.title("Top 10 Products")
plt.show()
```
